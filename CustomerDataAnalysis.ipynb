```python
#8. to update columns names and display teh updated column names
customers_df.rename(columns={'MntWines': 'Wines', 'MntFruits': 'Fruits', 'MntMeatProducts': 'Meat', 'MntFishProducts': 'Fish',
                   'MntSweetProducts': 'Sweet','MntGoldProds': 'Gold'}, inplace=True)
customers_df.columns.values.tolist()
```

```python
#9. Rename columns by removing 'Num' and 'Purchases'
customers_df.columns = customers_df.columns.str.replace('Num', '').str.replace('Purchases', '')

# Display the DataFrame to verify changes
customers_df.columns.values.tolist()
```

```python
#10. Keep only NumDealsPurchases column in the Promotion section
promotion_columns = [col for col in customers_df.columns if 'Deals' in col]
customers_df = customers_df[promotion_columns + [col for col in customers_df.columns if 'Deals' not in col]]

customers_df
```

```python
#11. Delete the column NumWebVisitsMonth
customers_df = customers_df.drop(columns=['WebVisitsMonth'])
customers_df.columns.values.tolist()
```

```python
#12. Create a new derived column named Age
current_year = 2024
customers_df['Age'] = current_year - customers_df['Year_Birth']
customers_df
```

```python
#13. Create a new column named Spending that holds the total amount spent on all products categories
product_columns = ['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']
customers_df['Spending'] = customers_df[product_columns].sum(axis=1)
customers_df
```

```python
#14. Inspecting the first few values of the Dt_Customer column to determine the format
print(customers_df['Dt_Customer'].head())

# Change the Dt_Customer type from object to datetime using the correct format
customers_df['Dt_Customer'] = pd.to_datetime(customers_df['Dt_Customer'], format='%d-%m-%Y')
```

```python
#15. Show the unique values of the column data Marital_Status
unique_marital_status = customers_df['Marital_Status'].unique()
print(unique_marital_status)
```

```python
#16. Classify customers in two segments for their marital status
customers_df['Marital_Status'] = customers_df['Marital_Status'].replace({
    'Divorced': 'Alone',
    'Single': 'Alone',
    'Absurd': 'Alone',
    'Widow': 'Alone',
    'Married': 'Not Alone',
    'Together': 'Not Alone'
})

# Display the unique values after transformation
updated_unique_marital_status = customers_df['Marital_Status'].unique()

# Display results to verify changes
unique_marital_status, updated_unique_marital_status, customers_df.head()
```

```python
#17. Show the unique values after transformation
updated_unique_marital_status = customers_df['Marital_Status'].unique()
# Show the unique values of the column Education
unique_education = customers_df['Education'].unique()
unique_education
```

```python
#18. Segment customers based on their education level
customers_df['Education'] = customers_df['Education'].replace({
    'Basic': 'Undergraduate',
    '2n Cycle': 'Undergraduate',
    'Graduation': 'Postgraduate',
    'Master': 'Postgraduate',
    'PhD': 'Postgraduate'
})

customers_df
```

```python
#19. Investigate and remove outliers in the Income column
# Calculate the IQR for Income
Q1 = customers_df['Income'].quantile(0.25)
Q3 = customers_df['Income'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Count the number of outliers before filtering
initial_outliers_count = len(customers_df[(customers_df['Income'] < lower_bound) | (customers_df['Income'] > upper_bound)])

# Filter out the outliers
customers_df = customers_df[(customers_df['Income'] >= lower_bound) & (customers_df['Income'] <= upper_bound)]

# Print the number of outliers removed
print("Number of outliers removed:", initial_outliers_count)

```

```python
from datetime import datetime

# Create a new DataFrame with the required columns
df_new = pd.DataFrame()

# Calculate the Age from Year_Birth
df_new['Age'] = datetime.now().year - df['Year_Birth']

# Calculate Spending as the sum of all spending columns
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df_new['Spending'] = df[spending_columns].sum(axis=1)

# Add the other required columns
df_new['Marital_Status'] = df['Marital_Status']
df_new['Education'] = df['Education']
df_new['Income'] = df['Income']
df_new['Children'] = df['Kidhome'] + df['Teenhome']
df_new['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df_new['Has_Complaint'] = df['Complain']

# Display the first few rows of the new DataFrame
df_new.head()
```

```python
import matplotlib.pyplot as plt

# Create a histogram for the number of children per customer
plt.figure(figsize=(8, 6))
plt.hist(df_new['Children'], bins=range(0, df_new['Children'].max() + 2), color='skyblue', edgecolor='black')
plt.title('Histogram of Number of Children per Customer')
plt.xlabel('Number of Children')
plt.ylabel('Number of Customers')
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# Histogram for Income
plt.figure(figsize=(8, 6))
plt.hist(df['Income'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of Income')
plt.xlabel('Income')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.show()

# Boxplot for Income
plt.figure(figsize=(8, 6))
plt.boxplot(df['Income'].dropna(), vert=False)
plt.title('Boxplot of Income')
plt.xlabel('Income')
plt.grid(axis='x')
plt.show()
```

```python
# Calculate the total spending per customer
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Spending'] = df[spending_columns].sum(axis=1)

# Histogram for Spending
plt.figure(figsize=(8, 6))
plt.hist(df['Spending'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Histogram of Spending')
plt.xlabel('Spending')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.show()

# Boxplot for Spending
plt.figure(figsize=(8, 6))
plt.boxplot(df['Spending'].dropna(), vert=False)
plt.title('Boxplot of Spending')
plt.xlabel('Spending')
plt.grid(axis='x')
plt.show()
```
